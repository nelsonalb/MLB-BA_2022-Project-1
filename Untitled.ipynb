```python
# Import the Data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load the Data
df_mlb = pd.read_csv('C:/Users/nelso/Documents/Project_1-Group_3/2022 MLB_Batting/Resources/Splits Leaderboard Data .csv')
df_mlb_lhp = pd.read_csv('C:/Users/nelso/Documents/Project_1-Group_3/2022 MLB_Batting/Resources/Splits Leaderboard Data vs LHP.csv')
df_mlb_rhp = pd.read_csv('C:/Users/nelso/Documents/Project_1-Group_3/2022 MLB_Batting/Resources/Splits Leaderboard Data vs RHP.csv')
```

```python
# Explore the Data
df_mlb.head()
```

```python
# Combine the data into a single dataset
df_mlb_complete = pd.merge(df_mlb_rhp, df_mlb_lhp, how="left", on=["Tm", "Tm"])
df_mlb_complete.head()
```

```python
# Combine the data into a single dataset
df_mlb_final = pd.merge(df_mlb_complete, df_mlb, how="right", on=["Tm", "Tm"])
df_mlb_final.head()
```

```python
# Save the merged dataset to a CSV file
df_mlb_final.to_csv('C:/Users/nelso/Documents/Project_1-Group_3/2022 MLB_Batting/Resources/df_mlb_final.csv', index=False)
```

```python
# Find the highest team batting average
# Group data by team and calculate batting average for each team
team_stats = df_mlb_final.groupby('Tm')[['H', 'AB']].sum()
team_stats['Avg'] = team_stats['H'] / team_stats['AB']

# Sort data by batting average
sorted_data = team_stats.sort_values(by='Avg', ascending=False)

# Select team with the highest batting average
highest_avg_team = sorted_data.iloc[0].name
highest_avg = sorted_data.iloc[0]['Avg']

print('Team with the highest batting average:', highest_avg_team)
print('Highest batting average:', highest_avg)
```

```python
# Calculate the batting average vs left and right handed pitchers
# Calculate batting average vs left and right handed pitchers
avg_vs_left = df_mlb_final['H_LHP'].sum() / df_mlb_final['AB_LHP'].sum()
avg_vs_right = df_mlb_final['H_RHP'].sum() / df_mlb_final['AB_RHP'].sum()

print('Batting average vs left handed pitchers:', avg_vs_left) 
print('Batting average vs right handed pitchers:', avg_vs_right)
```

```python
# Find the highest team batting average vs left and right handed pitchers
# Group data by team and calculate batting average for each team against left handed pitchers
team_stats_lhp = df_mlb_final.groupby('Tm')[['H_LHP', 'AB_LHP']].sum()
team_stats_lhp['Avg_vs_LHP'] = team_stats_lhp['H_LHP'] / team_stats_lhp['AB_LHP']

# Sort data by batting average vs left handed pitchers
sorted_data_lhp = team_stats_lhp.sort_values(by='Avg_vs_LHP', ascending=False)

# Select team with the highest batting average vs left handed pitchers
highest_avg_team_lhp = sorted_data_lhp.iloc[0].name
highest_avg_lhp = sorted_data_lhp.iloc[0]['Avg_vs_LHP']

# Find the highest team batting average
print('Highest team batting average vs left handed pitchers:', highest_avg_team_lhp)
print('Highest batting average vs left handed pitchers:', highest_avg_lhp)

# Group data by team and calculate batting average for each team against right handed pitchers
team_stats_rhp = df_mlb_final.groupby('Tm')[['H_RHP', 'AB_RHP']].sum()
team_stats_rhp['Avg_vs_RHP'] = team_stats_rhp['H_RHP'] / team_stats_rhp['AB_RHP']

# Sort data by batting average vs right handed pitchers
sorted_data_rhp = team_stats_rhp.sort_values(by='Avg_vs_RHP', ascending=False)

# Select team with the highest batting average vs right handed pitchers
highest_avg_team_rhp = sorted_data_rhp.iloc[0].name
highest_avg_rhp = sorted_data_rhp.iloc[0]['Avg_vs_RHP']

print('Highest team batting average vs right handed pitchers:', highest_avg_team_rhp)
print('Highest batting average vs right handed pitchers:', highest_avg_rhp)
```

```python
# Calculate the OBP for each team
team_stats['BB'] = df_mlb_final.groupby('Tm')['BB'].sum()
team_stats['HBP'] = df_mlb_final.groupby('Tm')['HBP'].sum()
team_stats['SF'] = df_mlb_final.groupby('Tm')['SF'].sum()
team_stats['OBP'] = (team_stats['H'] + team_stats['BB'] + team_stats['HBP']) / (team_stats['AB'] + team_stats['BB'] + team_stats['HBP'] + team_stats['SF'])

# Calculate the runs scored per game for each team
team_stats['Runs/G'] = df_mlb_final.groupby('Tm')['R'].mean() / 162

print(team_stats)
```

```python
# Find the highest team runs
max_runs_team = team_stats['Runs/G'].idxmax()
team_stats.sort_values(by='Runs/G', ascending=False).iloc[0]['Runs/G']
print (max_runs_team)
```

```python
# Correlation between a team's on-base percentage (OBP) and its runs scored per game. This could provide insight into how effective a team is at getting players on base and turning those opportunities into runs
# Plot the correlation between OBP and Runs/G
plt.scatter(team_stats['OBP'], team_stats['Runs/G'])
plt.title('Correlation between OBP and Runs Scored per Game')
plt.xlabel('On-Base Percentage (OBP)')
plt.ylabel('Runs Scored per Game')

# Add the equation to the plot
m, b = np.polyfit(team_stats['OBP'], team_stats['Runs/G'], 1)
plt.annotate(f'y = {m:.2f}x + {b:.2f}', (0.3, 5.15))

plt.show()
```

```python
# Correlation between a team's on-base percentage (OBP) and its runs scored per game. 
sns.regplot(x=team_stats['OBP'], y=team_stats['Runs/G'])
plt.title('Correlation between OBP and Runs Scored per Game')
plt.xlabel('On-Base Percentage (OBP)')
plt.ylabel('Runs Scored per Game')
plt.show()

from scipy.stats import pearsonr

corr, _ = pearsonr(team_stats['OBP'], team_stats['Runs/G'])
print(f'Correlation: {corr:.2f}')
```

**Discussion about the linear relationship:** The scatter plot shows that there is a positive correlation between OBP and Runs/G, indicating that teams with higher OBP tend to score more runs per game. The regression line shows the overall trend of the data, with a positive slope indicating a positive correlation.

```python
# Calculate the correlation coefficient between BA and Runs/G
corr = np.corrcoef(team_stats['Avg'], team_stats['Runs/G'])[0, 1]
print(f"Correlation coefficient between Avg and Runs/G: {corr:.2f}")

# Plot the correlation between BA and Runs/G
plt.scatter(team_stats['Avg'], team_stats['Runs/G'])
plt.title('Correlation between Batting Average and Runs Scored per Game')
plt.xlabel('Batting Average (BA)')
plt.ylabel('Runs Scored per Game')

# Add the equation to the plot
m, b = np.polyfit(team_stats['Avg'], team_stats['Runs/G'], 1)
plt.annotate(f'y = {m:.2f}x + {b:.2f}', (0.225, 5.15))

# Add the linear regression line to the plot
m, b = np.polyfit(team_stats['Avg'], team_stats['Runs/G'], 1)
plt.plot(team_stats['Avg'], m*team_stats['Avg'] + b, color='red')

plt.show()
```

**Discussion about the linear relationship:** The correlation coefficient between BA and Runs/G is positive, indicating a moderately strong correlation between the two variables.

The scatter plot shows a generally upward trend, suggesting that teams with higher batting averages tend to score more runs per game.
